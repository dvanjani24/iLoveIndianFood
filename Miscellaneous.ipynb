```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

with open("List of Indian Foods.txt", "r") as file:
    text = file.read()
```

```python
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
plt.figure(figsize=(15, 15))
# Add a red theme to the word cloud
wordcloud.recolor(colormap="Reds_r")
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import load_img, img_to_array

# Load the image
image_path = "Indian Food Images (Train)/Indian Food Images/kadai_paneer/Image_2.jpg"
image = load_img(image_path)

# Convert the image to a numpy array
image_array = img_to_array(image)

# Create an instance of ImageDataGenerator with shear and zoom transformations
datagen = ImageDataGenerator(shear_range=0.4, zoom_range=0.2, vertical_flip=True)

# Reshape the image array to have a batch dimension
image_array = image_array.reshape((1, *image_array.shape))

# Generate augmented images
augmented_images = datagen.flow(image_array, batch_size=1)

# Get the original and augmented images
original_image = image_array[0]
augmented_image = augmented_images.next()[0]

# Display the before and after images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_image.astype(int))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(augmented_image.astype(int))
plt.title('Augmented Image')
plt.axis('off')


plt.tight_layout()
plt.show()
```
