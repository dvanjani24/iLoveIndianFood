```python
import pickle
file_path = 'train_history/history_resnet101'
with open(file_path, 'rb') as file:
    history_resnet101 = pickle.load(file)
    
file_path = 'train_history/history_inception_resnet'
with open(file_path, 'rb') as file:
    history_inception_resnet = pickle.load(file)
    
file_path = 'train_history/history_mobilenet'
with open(file_path, 'rb') as file:
    history_mobilenet = pickle.load(file)
    
file_path = 'train_history/history_mobilenet_ft'
with open(file_path, 'rb') as file:
    history_mobilenet_ft = pickle.load(file)
    
file_path = 'train_history/my_cnn'
with open(file_path, 'rb') as file:
    my_cnn = pickle.load(file)
```

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

def plot_training_history(history, training_color = "black", validation_color = "#800000"):
    # Access the training history data
    loss = history['loss']
    accuracy = history['accuracy']
    val_loss = history['val_loss']
    val_accuracy = history['val_accuracy']

    # Set the color scheme
    training_color = 'black'
    validation_color = '#800000'
    # Increase font size
    plt.rcParams.update({'font.size': 16})
    # Plot the training and validation loss
    plt.figure(figsize=(8, 6))
    plt.plot(loss, color=training_color, label='Training Loss')
    plt.plot(val_loss, color=validation_color, label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plot the training and validation accuracy
    plt.figure(figsize=(8, 6))
    plt.plot(accuracy, color=training_color, label='Training Accuracy')
    plt.plot(val_accuracy, color=validation_color, label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_training_history(history_resnet101)
```

```python
plot_training_history(history_inception_resnet)
```

```python
plot_training_history(history_mobilenet)
```

```python
plot_training_history(history_mobilenet_ft)
```

```python
plot_training_history(my_cnn)
```
